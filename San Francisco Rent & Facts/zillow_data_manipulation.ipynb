```python
import pandas as pd
```

```python
df = pd.read_csv("Zip_zori_uc_sfrcondomfr_sm_month.csv")
df.head()
```

```python
# Setting multiple columns as the index
df = df.set_index(['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'City', 'CountyName'])

# Stacking the DataFrame
stacked_df = df.stack()

# Resetting index after stacking to flatten the DataFrame
stacked_df = stacked_df.reset_index()

# Drop rows where the column values are of type 'object' (e.g., strings)
stacked_df = stacked_df[stacked_df[0].apply(lambda x: not isinstance(x, str))]

stacked_df.head()
```

```python
# Assuming 'level_0', 'level_1', ..., are the names you want to change
stacked_df = stacked_df.rename(columns={
    'level_8': 'YearMonth',
    0: 'Rent',
    # continue for other columns you wish to rename
})

stacked_df.head()
```

```python
stacked_df.info()
```

```python
# Extract year and month using string slicing
stacked_df['Year'] = stacked_df['YearMonth'].str[:4].astype(int)  # Convert to int for year
stacked_df['Month'] = stacked_df['YearMonth'].str[5:7].astype(int)  # Convert to int for month

stacked_df.head()
```

```python
df2 = pd.read_excel("sf_demo_zip.xlsx")
df2.head()
```

```python
df3 = pd.merge(stacked_df, df2, left_on='RegionName',right_on='zipcode',how='inner')

df3[df3['City']=='San Francisco'].head()
```

```python
df3 = df3.drop('zipcode', axis=1)  # axis=1 indicates column
df3 = df3.fillna(0)
df3.head()
```

```python
df3.info()
```

```python
df4 = df3[df3['City'] == 'San Francisco']
```

```python
# Export DataFrame to a CSV file
df4.to_csv('zillow_us_rent.csv', index=False)
```
